<h1>Object  Reconstruction Examples</h1>

<h2> 1. Object Prediction - Touch Only Training </h2>

Here we train a touch only object prediction model in the poking setting. 

```python
import os
from pterotactyl.reconstruction import touch 
from pterotactyl.reconstruction.vision import train as vision_train 
from pterotactyl import pretrained

TOUCH_LOCATION = os.path.dirname(pretrained.__file__) + '/reconstruction/touch/best/'

class Params: # define training arguments 
    def __init__(self):
        self.cut = 0.33
        self.seed = 0 
        self.limit_data = True
        self.epochs = 30
        self.lr = 0.0003
        self.eval = False
        self.batch_size = 16
        self.number_points = 30000
        self.loss_coeff = 9000
        self.exp_id = "example"
        self.exp_type = "t_p_example"
        self.patience = 10
        self.visualize = False
        self.finger = True 
        self.num_grasps = 5 
        self.hidden_GCN_size = 300
        self.num_GCN_layers = 20 
        self.CNN_ker_size = 5 
        self.layers_per_block = 3 
        self.num_CNN_blocks = 6
        self.use_touch = True # using touch 
        self.use_img = False # not using vision 
        self.val_grasps = -1
        self.touch_location = TOUCH_LOCATION 
        
        
    

params = Params()
vision_trainer = vision_train.Engine(params)

vision_trainer()
```

<h2> 2. Object Prediction - Vision and Touch Training </h2>

Here we train a vision and touch object prediction model in the grasping setting. 

```python
params.finger = False 
params.use_img = True 
vision_trainer = vision_train.Engine(params)
vision_trainer()
```

<h2> 3. Object Prediction - Touch only Evaluation </h2>

Here we evaluate the pretrained touch only reconstruction model in the poking setting. This will provide the reconstruction score on the test set. Here visualized is set to true so that predicted visualization can be observed. 

```python
params.finger = True 
params.use_img = False 
params.eval = True 
params.pretrained = False
params.visualize = True 
params.exp_type = "t_p_example"
params.exp_id = "example"
        

vision_trainer = vision_train.Engine(params)
vision_trainer()
```

As we set the visualization parameter to True we can now visualize a few predictions. 

```python
import os 

from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw 
from glob import glob 

from pterotactyl import objects
FONT_LOCATION = os.path.dirname(objects.__file__) + '/OpenSans-Bold.ttf'


img_locations = glob('results/t_p_example/example/*/')
for location in img_locations[:5]:
    mesh = Image.open(location + "mesh.png")
    points = Image.open(location + "points.png")
    gt = Image.open(location + "ground_truth.png")
    
    
    combo = Image.new('RGB', (512*3, 512))
    x_offset = 0
    for im, st in [[gt, 'Ground Truth'], [mesh, 'Mesh'], [points, 'Point Cloud']]:
        draw = ImageDraw.Draw(im)
        font = ImageFont.truetype(FONT_LOCATION, 35)
        W,H = im.size
        w,h = font.getsize(st)
        draw.text(((W-w)/2, 0),st,(0, 0, 0), font = font)
        combo.paste(im, (x_offset, 0))
        x_offset += im.size[0]
    display(combo)

```
